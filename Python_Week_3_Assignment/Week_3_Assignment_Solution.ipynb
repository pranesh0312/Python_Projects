# 5)Convert datatype of column published_at from object to pandas datetime.

# 6. Create a new column as 'published_month' using the column published_at (display the months only)

```python
#Inserting new column 'published_month'

data_file['published_month']=data_file['published_at'].dt.month
```

dt.month--->used to extract month from the datetime format

```python
#Dataset after inserting the new column
data_file
```

# 7) Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'and so on.....

```python
#Replacing the published_month with the month name using map() functioin

month_value = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr',
    5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug',
    9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
data_file['published_month'] = data_file['published_month'].map(month_value)
```

map()-->used to finds a match for the column value in the dictionary 

```python
#Dataset after replacing the published_month column with the month name
data_file
```

# 8) Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

# 13)Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
data_file['view_count'].corr(data_file['dislikes'])
```

```python
sns.scatterplot(x='view_count', y='dislikes', data=data_file)
plt.title('no.of Views vs Dislikes')
plt.xlabel('Number of Views')
plt.ylabel('Number of Dislikes')
plt.show()
```

From the correlation value(0.6847595531934572),it shows that the correlation is positive(i.e The number of views have an effect on how many people disliked the video)

From the scatterplot,it discribes that there is an increase in the pattern,this shows that when the number of views increases and the number of dislikes also increases
