### <font color='blue'>Parking Lot Service Module</font>
<p>Implementation of Abstract behaviour using park_vehicle, unpark_vehicle & company_parked_vehicles</p>

```python
from abc import ABC, abstractmethod
```

```python
class ParkingLotService(ABC):
    
    @abstractmethod
    def park_vehicle(self, vehicle):
        pass
    
    @abstractmethod
    def unpark_vehicle(self, vehicle_plate_number):
        pass
    
    @abstractmethod
    def company_parked_vehicles(self, company_name):
        pass
```
