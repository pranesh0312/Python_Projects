### Pyplot module supports variety of plotting functions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Week4_Profile_Data.csv')
df.info()
```

```python
df.head()
```

#### Line Plot

```python
sns.lineplot(x=df["Name"], y=df["Age"], data=df)
plt.show()
```

#### Bar plot

```python
a=sns.barplot(df,x='Name',y='Age')
for i in a.containers:
    a.bar_label(i,)
plt.show()
```

#### Histogram

```python
random_num = np.random.randn(100)
sns.histplot(random_num)
plt.show()
```

#### Scatter Plot

```python
sns.scatterplot(x=df["Name"], y=df["Age"], data=df)
plt.show()
```

## Seaborn

Build on top of Matplotlib

#### Displot

To draw distribution plots, seaborn uses interfaces of figure level

```python
sns.displot(random_num,kde=True)
plt.show()
```

```python
penguins_data = sns.load_dataset("penguins")
sns.displot(data=penguins_data, x="flipper_length_mm", kde=True) 
plt.show()
```

#### Joint plot

It draws plot of two variables along with univariate and bivariate plots

```python
sns.jointplot(data=penguins_data, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.show()
```

```python
sns.jointplot(data=penguins_data, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
plt.show()  #kde - defines which approch to choose, hue- defines the color of plot elements
```

#### Heat map

This represnets a color-encoded matrix

```python
corre=penguins_data.corr(numeric_only=True)
```

```python
sns.heatmap(corre, cbar=0, linewidths=2,vmax=1, vmin=0, square=True, cmap='Blues', annot = True)
plt.show()
```

#### Count plot

```python
cx = sns.countplot(x="body_mass_g",data=penguins_data)
for i in cx.containers:
    cx.bar_label(i,)
plt.show()
```

#### Pair plot

Shows pairwise relationships in data

```python
sns.pairplot(penguins_data, hue='species',kind='scatter',palette='husl')
plt.show()
```

#### Swarm plot

This represents a scatterplot which is categorical in nature

```python
sns.set_theme(style="whitegrid") # style - a preconfigured style is selected from a dictionary
tips_data = sns.load_dataset("tips")
ax_grid = sns.swarmplot(x=tips_data["total_bill"]) 
```

Categorial values can be used to group the swarms

```python
ax_grid = sns.swarmplot(x="day", y="total_bill", data=tips_data) 
```
